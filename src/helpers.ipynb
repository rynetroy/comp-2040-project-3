```python
import pandas as pd

# Replace spaces with underscores
def clean_column_names(df):
    """Standardize column names."""
    df.columns = df.columns.str.lower().str.replace(" ", "_") 
    return df
```

```python
# Create a new column 'is_closed' that is True
# if 'status' is in the closed_status list, otherwise False
def create_is_closed(df):
    """Create closed business indicator."""
    closed_status = ["Closed (L)", "Ceased Operation", "Cancelled"]
    df["is_closed"] = df["status"].isin(closed_status)
    return df
```

```python
def filter_downtown(df):
    """Filter to downtown records."""
    return df[df["community_characterization_area"] == "Downtown"]
```

```python
# new helper function added for week 14
def assign_phase(year):
    """Assign a development phase label based on start year."""
    if year < 2015:
        return 'Pre-2015 establishment'
    elif year <= 2020:
        return '2015\u20132020 transition'
    else:
        return '2020+ restructuring'

print("assign_phase(2012) ->", assign_phase(2012))
print("assign_phase(2018) ->", assign_phase(2018))
print("assign_phase(2023) ->", assign_phase(2023))
```
